# Matplotlib example workflow

```python
#0. import matplotlib and get it ready for plotting in jupyter
import matplotlib.pyplot as plt
plt.close('all')
#1. prepare the data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

#2. setup the plot
fig, ax = plt.subplots(figsize=(10,10))

#3. Plot data
ax.plot(x, y)

#4. Customize plot
ax.set(title='Simple plot',
       xlabel='eixo-x',
       ylabel='eixo-y')

#5. Save & show
fig.savefig("images/sample-plot.png")
plt.show()
```

```python
# Plot from dictionary
nut_butter_price = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_price.keys(), nut_butter_price.values())
ax.set(title="Lucas's Nut Butter Store",
      ylabel='Price($)')
plt.show();
```

```python
fig, ax = plt.subplots()
ax.barh(nut_butter_price.keys(), nut_butter_price.values());
plt.show();
```

```python
# Make some data
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
plt.show();
```

### Two options for subplot

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                              ncols=2,
                                              figsize=(10, 5))
# plot to each different axis
ax1.plot(x, np.tan(x))
ax2.scatter(x, np.tan(x))
ax3.bar(nut_butter_price.keys(), nut_butter_price.values())
ax4.hist(np.sin(x))
plt.show();
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))
# Plot to each different index
# plot to each different axis
ax[0, 0].plot(x, np.tan(x))
ax[0, 1].scatter(x, np.tan(x))
ax[1, 0].bar(nut_butter_price.keys(), nut_butter_price.values())
ax[1, 1].hist(np.sin(x))

plt.show()
```

# Which one should you use? (pyplot vs matplotlib OO method)
* When plotting something quiclky, can use the pyplot method
* When plotting something more advanced, use the OO method

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease['age'] >= 50]
len(over_50), len(heart_disease)
```

```python
#Pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
            figsize=(10, 6))
plt.show()
```

```python
#OO method mixed with pyplot
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax)
#ax.set_xlim([45, 100])
plt.show()
```

```python
#OO method
fig, ax = plt.subplots(figsize=(10, 6))

#Plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
#Customize
ax.set(title='Heart Disease and Cholesterol levels',
       xlabel='Age',
       ylabel='Chol')

#add a legend
ax.legend(*scatter.legend_elements(), title='Target')

#add a horizontal line
ax.axhline(over_50['chol'].mean(), 
           linestyle='--',
           color='red');
plt.show()
```

```python
over_50.head()
```

```python
# sub plot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10))

# add a data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'])
#customize ax0
ax0.set(title='Heart disease and Cholesterol levels',
        xlabel='age',
        ylabel='chol')
# legend ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# add a line
ax0.axhline(y=over_50['chol'].mean(),
           linestyle='--',
           color='red')


# add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'])
#customize ax1
ax1.set(title='Heart disease and Max Heart hate',
        xlabel='age',
        ylabel='max heart hate')
#legend ax1
ax1.legend(*scatter.legend_elements(), title='Target')

#add a line
ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--',
            color='red')

plt.show()
```

#### Most clean graph

```python
# sub plot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10))

# add a data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'])
#customize ax0
ax0.set(title='Heart disease and Cholesterol levels',
        ylabel='chol')
# legend ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# add a line
ax0.axhline(y=over_50['chol'].mean(),
           linestyle='--',
           color='red')


# add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'])
#customize ax1
ax1.set(title='Heart disease and Max Heart hate',
        xlabel='age',
        ylabel='max heart hate')
#legend ax1
ax1.legend(*scatter.legend_elements(), title='Target')

#add a line
ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--',
            color='red')
#add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')

#save a fig
fig.savefig("images/heart-disease-analysis.png")
plt.show()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
# sub plot of chol, age, thalach
fig, ax0 = plt.subplots(nrows=1,
                              ncols=1,
                              figsize=(10,10))

# add a data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap='winter') # This change the colour scheme
#customize ax0
ax0.set(title='Heart disease and Cholesterol levels',
        ylabel='chol')
# legend ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# add a line
ax0.axhline(y=over_50['chol'].mean(),
           linestyle='--',
           color='red')
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
# sub plot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)

# add a data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap='winter')
#customize ax0
ax0.set(title='Heart disease and Cholesterol levels',
        ylabel='chol')

# Change the x axis limits
ax0.set_xlim([45, 80])
# change the y axis limits
ax0.set_ylim([0, 600])
# legend ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# add a line
ax0.axhline(y=over_50['chol'].mean(),
           linestyle='--',
           color='red')

# add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                      cmap='winter')
#customize ax1
ax1.set(title='Heart disease and Max Heart hate',
        xlabel='age',
        ylabel='max heart hate')

#legend ax1
ax1.legend(*scatter.legend_elements(), title='Target')

# Change the x axis limits
ax1.set_xlim([45, 80])
# change the y axis limits
ax1.set_ylim([0, 200])

#add a line
ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--',
            color='red')
#add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
