### Tuning the hyperparamets

```python
# first we go to create a function to show the accuracy score of a model
def evaluate_model(models, X, y):
    """
    Evaluate the accuracy of models.

    Args:
        models (dict): Dictionary with model names as keys and model instances as values.
        X (array): Feature data (no labels).
        y (array): Data labels.

    Returns:
        dict: Dictionary with the scores of the models.
    """

    evaluated_models = {}
    for name, model in models.items():
        # training using the cross_val_score
        cv_acc = cross_val_score(model, X, y, cv=5, scoring='accuracy')
        # save the result in evaluated_models
        evaluated_models[name] = np.mean(cv_acc)

    return evaluated_models
```

```python
# set a seed
np.random.seed(7)
# Create a dict with the model
models = {"Linear SVC": LinearSVC(),
          "Neighbours Classifier": KNeighborsClassifier(),
          "Random Forest Classifier": RandomForestClassifier()}

# evaluate the model
evaluated_models = evaluate_model(models, X, y)
# transform the dict python into pandas' dataframe
evaluated_models =  pd.DataFrame(evaluated_models, index=['accuracy'])
evaluated_models
```

```python
evaluated_models.T.plot(kind='bar');
```

When we use the `cross_val_score` we can to train and test in all datas, so our accuracy are changed!
So we need to hyperparamter our model!

```python
import time

# Using randomizedSearchCV to tuning auto
def tuning_model(models, param_distributions, X, y):
    """
    Tuning auto the model
    
    Args:
        models (dict): Dictionary with model names as keys and model instances as values.
        param_distributions (dict): with the params to test the model
        X (array): Feature data (no labels).
        y (array): Data labels.

    Returns:
        dict: Dictionary with the name, the best paramets and scores of the models.
    """
    # set the random seed
    np.random.seed(7)

    # a dict to storage the best models
    best_models = {}

    # tuning each model of dict
    for name, model in models.items():
        # find the model and our model
        if name in param_distributions:
            param_grid = param_distributions[name]

        # define the hyperparametrs
        model_test = RandomizedSearchCV(model,
                                  param_distributions=param_grid,
                                  cv=5,
                                  n_iter=10,
                                  scoring='accuracy');
        # train the model and get the time spend
        start = time.time()
        model_test.fit(X, y)
        end = time.time()
        
        # save the best model
        best_models[name] = {"best_params": model_test.best_params_,
                            "best_score": model_test.best_score_
        }
        
        print(f'The model {name} was trained in: {(end - start):.2f} sec')
        
    # return the dict
    return best_models
```

```python
# Create a dict with the model
models = {"Linear SVC": LinearSVC(),
          "Multinomial NB": MultinomialNB(),
          "Gaussian NB": GaussianNB(),
          "Decision Tree Classifier": DecisionTreeClassifier(),
          "MLPClassifier": MLPClassifier(),
          "Ada Boost Classifier": AdaBoostClassifier(),
          "Neighbours Classifier": KNeighborsClassifier(),
          "Gradient Boosting Classifier": GradientBoostingClassifier(),
          "Random Forest Classifier": RandomForestClassifier(),}

# Creating dicts to params to the model
linear_svc_grid = {
    'C': np.arange(1, 10, 0.5),
    'max_iter': np.arange(100, 2000, 100),
}

knn_grid = {
    'n_neighbors': np.arange(1, 31, 1),
}

gbc_grid = {
    'n_estimators': np.arange(50, 501, 50),
    'learning_rate': np.linspace(0.01, 0.3, 10),
    'max_depth': np.arange(1, 11, 1)
}

rfc_grid = {
    'n_estimators': np.arange(50, 501, 50),
    'max_features': ['sqrt', 'log2', None] + list(np.arange(1, 5, 1)),
    'max_depth': [None] + list(np.arange(5, 51, 5)),
    'min_samples_split': np.arange(2, 10, 1),
    'min_samples_leaf': np.arange(1, 10, 1),
    'n_jobs': [-1],
}

gnb_grid = {
    'var_smoothing': np.logspace(0, -18, 1000)
}

mnb_grid = {
    'alpha': np.arange(0, 1000000, 0.01),
    'fit_prior': [True, False],
}

dtc_grid = {
    'splitter': ['best', 'random'],
    'max_depth': [None] + list(np.arange(5, 51, 5)),
    'min_samples_split': np.arange(2, 10, 1),
    'min_samples_leaf': np.arange(1, 10, 1),
    'max_features': ['sqrt', 'log2', None] + list(np.arange(1, 5, 1)),
}

mpc_grid = {
    'alpha': np.linspace(0.0001, 1, 10),
    'learning_rate': ['constant', 'invscaling', 'adaptive'],
    'hidden_layer_sizes': [(50,), (100,), (50, 50)], 
    'activation': ['identity', 'logistic', 'tanh', 'relu']
}

abc_grid = {
    'n_estimators': np.arange(50, 501, 50),
    'learning_rate': np.arange(0.1, 10.1, 0.5),
}

# create a dict with the all paramets to the model
param_distributions = {"Linear SVC": linear_svc_grid,
                       "Neighbours Classifier": knn_grid,
                       "Gradient Boosting Classifier": gbc_grid,
                       "Random Forest Classifier": rfc_grid,
                       "Multinomial NB":mnb_grid,
                       "Gaussian NB": gnb_grid,
                       "Decision Tree Classifier": dtc_grid,
                       "MLPClassifier": mpc_grid,
                       "Ada Boost Classifier": abc_grid}
```
